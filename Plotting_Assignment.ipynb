````python
# Creating a heatmap of the data (without annotation and colormap)
plt.figure(figsize=(10, 4)) #(10, 4) ---> (width, length)
columns = ["year", "cylinders", "displ", "pv2", "pv4", "city", "UCity", "highway", "UHighway", "comb", "co2", "feScore", "ghgScore"]

df1 = df[columns]

# to show how closely related each column is
corre = df1.corr()

# annotation set to true shows the value of the correlaton for analysis
# vmin and vmax are parameters to specify the data range of values for the colormap
sns.heatmap(corre)
plt.title("Heatmap")
plt.show()
````

````python
# Creating a heatmap of the data (with annotation and colormap)
plt.figure(figsize=(10, 4)) #(10, 4) ---> (width, length)
columns = ["year", "cylinders", "displ", "pv2", "pv4", "city", "UCity", "highway", "UHighway", "comb", "co2", "feScore", "ghgScore"]

df1 = df[columns]

# to show how closely related each column is
corre = df1.corr()

# annotation set to true shows the value of the correlaton for analysis
# vmin and vmax are parameters to specify the data range of values for the colormap
sns.heatmap(corre, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title("Heatmap")
plt.show()
````

## **Data Interpretation**

The heatmap Interpretation can be done based on Correlation Values or Colormap as seen in the image above.



> **Positive correlation values indicate a direct relationship (as one variable increases, so does the other), while negative values indicate an inverse relationship (as one variable increases, the other decreases). Values close to 1 or -1 indicate strong correlations, whereas values close to 0 indicate weak or no correlation.**



*Correlation Values Interpretation:*
*  **Cylinders and CO2 Emissions (0.85):** This strong positive correlation shows that vehicles with more cylinders tend to have higher CO2 emissions. This is expected, as engines with more cylinders usually consume more fuel, leading to increased CO2 emissions.

*  **Greenhouse Gas Score (ghgScore) and CO2 Emissions (-0.94):** The greenhouse gas score has a strong negative correlation with CO2 emissions. Higher CO2 emissions are associated with a lower greenhouse gas score, depicting the vehicle's larger environmental impact. With this info, we can say that ghgScore is negatively correlated, implying that both of them don't affect each other.

*  **Year:** The weak correlation of the year with most variables shows that there isn't a strong trend towards more or less fuel efficiency over the years in this dataset. I can also interpret year to be that the cars being brand new doesn't really affect fuel efficiency. With this info, we can say that year has a low correlation to fuel efficiency.

> **This ``` coolwarm ``` colormap visually shows the correlation values. It ranges from blue (indicating a strong negative correlation) to red (indicating a strong positive correlation).**

*Colormap Interpretation:*


1. **Blue Shades:**

* **Strong Negative Correlation (-1 to 0):** The blue shades represent negative correlations, with darker blues showing stronger negative relationships. For instance, displ (engine displacement) and feScore have a dark blue cell with a correlation value of -0.93, showing that as engine displacement increases, the fuel efficiency score decreases significantly.

2. **Red Shades:**

* **Strong Positive Correlation (0 to +1):** The red shades indicate positive correlations, with darker reds showing stronger positive relationships. For example, the correlation between co2 and cylinders is 0.85, highlighted in a dark red cell, meaning that as the number of cylinders increases, CO2 emissions also rise.

3. **Neutral (White) Areas:**

* **Weak or No Correlation (close to 0):** The cells with colours closer to white show weaker correlations, near zero, meaning little to no linear relationship between the variables. For example, the correlation between year and most other variables is close to 0, shown in lighter shades, depicting that the year does not significantly influence those variables in this dataset.
